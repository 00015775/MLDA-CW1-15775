## correlation matrix

````python
""" 
    The values range from -1 to 1(Pearson method), correlation coefficients closer to 1 will indicate positive relationship
    while those below 0 and close to -1, will indicate negative relationship. While 0 means there is no predictable or no
    linear relationship between these two variables (which could imply there could be a non-linear relationship).
    pattern/relationship between those two variables.
    As of now there are only 16 numerical columns
"""

corr_matrix = student_por_df.corr(numeric_only=True, method="pearson")
display(corr_matrix)
print(f"\nShape of correlation matrix: {corr_matrix.shape}\n")
````

````python
# to have a nice visualization of relationship of each variable in terms of colors

plt.figure(figsize=(17,12))
sns.heatmap(data=corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("\nCorrelation of all numerical columns.")
plt.show()
````

````python
"""
    Here is the triangular corelation matrix, that is each corelation coefficient is repeated only once, which is
    unlike in a just corelation matrix which is effectively is a symmetric matrix with the identical values in both above and below
    the diagonals(which are always 1).

    Since correlation of a variable with itself is always 1, their colors are not drawn below, since no valuable information
    can be learned from this redundancy.
"""

tri_mask = np.triu(np.ones_like(corr_matrix)) # entries above the diagonal are 1, below are 0

plt.figure(figsize=(18,10))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=True, fmt=".2f", mask=tri_mask)
plt.title("\nTriangular Correlation Heatmap of all(16) numerical columns.")
plt.savefig("plots/corr_triu_num_inputs.png")
plt.show()
````

```text
Based on corelation matrices, given that no data preprocessing has been done on categorical values to map them to numerical values, as of now providing corelation only for numerical columns(16), there is a high correlation of dependent value "G3" with "G2", "G1", with 0.91 and 0.82 coefficients respectively. After that we can start potentially removing input features that do not contribute much to the final output. But do note that 0 corelation coefficient does not imply that there is no relationship between two variables, since Pearson corelation only checks for the linear dependence/relationship between two values, the 0 coefficient could mean that there is a non-linear relationship between given variables. Overall, we will plot correlation heatmaps in the preceding Data Preprocessing section as well, in order to test the significance of corelation coefficient with any of 90, 95 and 99% confidence scores and help us to decide with feature selection process (filter method). Or use wrapper methods that includes using grid search approach that tends to give higher accuracy compared to manual filter method approach.

Since some of the categorical values are either nominal or ordinal, we can use Spearman rank correlation or Kendall methods for ordinal features which are available in pandas .corr() function, while for nominal data we can use either Phi coefficient or Cramer's V which is available in scipy.stats library. Or also use the biserial correlation, that is a correlation between on one ore more quantitative variables, and on the other hand one or more binary variables.
```

````python
fig, axes = plt.subplots(nrows=1, ncols=4, figsize=(16, 4))

for idx in range(4):
    col = student_por_df.columns[idx]
    data = student_por_df[col]

    bins = len(data.unique()) if data.min() != data.max() else 1

    axes[idx].hist(data, bins=bins, edgecolor="black")
    axes[idx].set_title(f"{col.capitalize()} frequency.")
    axes[idx].set_xlabel("Values")
    axes[idx].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
````

````python
# display frequency distribution for all 33 attributes

colors = ["darkred","crimson","orangered","orange","gold","yellow",
    "limegreen","green","darkgreen","teal","cyan","deepskyblue","dodgerblue","blue",
    "navy","darkviolet","purple","magenta","deeppink","brown",
    "saddlebrown","chocolate","peru","olive","darkgoldenrod","slategray","darkslategray",
    "black","dimgray","lightgray","gainsboro","lightcoral","tomato","lightsalmon" ]

num_cols = len(student_por_df.columns)   
rows = 9
cols = 4

fig, axes = plt.subplots(rows, cols, figsize=(18, 40))
axes = axes.flatten()  

for idx in range(num_cols):
    col = student_por_df.columns[idx]
    data = student_por_df[col]

    bins = len(data.unique()) if data.min() != data.max() else 1

    axes[idx].hist(data, bins=bins, edgecolor="black", color=colors[idx])
    axes[idx].set_title(f"{col.capitalize()} frequency.")
    axes[idx].set_xlabel("Values")
    axes[idx].set_ylabel("Frequency")

for k in range(num_cols, rows * cols):
    axes[k].set_visible(False)

plt.tight_layout(pad=1.3, h_pad=3)
plt.savefig("plots/frequency_dist_33_attr.png")
plt.show()
````

## feature selection

* As discussed [above](#correlation_does_not_mean_causation) for feature selection I was planning to use manual statistical method (which is also known as *filter method*) of applying Pearson/Kendall correlation coefficients to determine which features to use for predicting `G3` final output. However, after delving more into feature selection methods, I found out that while filter method being less computationally expensive and being relatively fast compared to other methods, it produces relatively lower accuracy compared to methods which somewhat decides by itself based on whether to include that particular features or not based on an increase of overall accuracy. This method is known as *wrapper method*.
  
* *Wrapper method* while being relatively better in terms of accuracy compared to *Filter methods*, it comes with its flaws, that is, it tends to overfit and generalize poorly to the unseen data, while also being more computationally expensive. Even if this project is not intended to be on the production level and seeing tons of new data that the model has not seen before, I reasoned that it is a still good practice to use another approach which produces better accuracy than *filter methods*, but does not overfit as much as *wrapper methods*.
  
* After delving even further, I found that there is another approach which combines the best of both worlds, by yielding high accuracy while not overfitting, that approach is called *embedded methods*. So for feature selection, *embedded methods* will be used, specifically **Lasso** which performs both feature selection and regularization as not to overfit, and other flavours of *embedded methods* may be used as well if needed.

<div class="alert alert-block alert-info", style="font-weight: 600">
        However, since both wrapper and embedded methods are model dependent (<u>model-based</u>), meaning it is applied only <i>during</i> model training because it is a type of feature selection approach that is built directly into the model building/training procedure rather than selecting features separately beforehand like filter methods, embedded methods perform feature selection at the same time as the model is being trained, by identifying which features contribute the most to the model's overall accuracy. <u>Hence, <i>Embedded methods</i> will be applied during model training process!</u>
</div>

## feature engineering

* Up until this points, most of the steps involed in feature engineering have been applied (*model-based techniques will be applied within the model training process*) which are feature transformation, selection, scaling, data cleaning, handling errors and encoding. However, the feature engineering also involes *feature creation*, that is creation of new features based on the existing ones within the dataset.

* However, based on the sources found, ***feature creation* is primarily based on the domain knowledge**, which comes from either industry knowledge or having dedicated experience in specific domain. The encoding steps applied here which are `one-hot-encoding` or `ordinal-encoding`, <u>are considered as a one of the *feature creation* techniques</u>, effectively creating new additional columns based on existing ones. But *feature creation* also involves creating new so-called *synthetic features*, and require domain expertise for correctly doing that, **for this project no *synthetic features* will be created**, that is raising any of the features to the power higher than 1, or multiplying existing features to get a new *synthetic features*, which tends to provide a model that looks like a polynominal function (*polynominal transforms*), such as $f(x) = a_nx^n + a_{n-1} x^{n-1} + \dots + a_1 x + a_0$. The reason being is lack of expertise knowledge in working with students performance related data and domain.

# Models Training


* As discussed above there has been minimal feature scaling done on the dataset due to the range of numerical features being more or less on the similar scale. Figuring that it could be an issue for certain type of models, specifically non-linear ones, I found out that **tree-based models** require minimal to **no feature scaling** required, which are `Decision Tree`, `Random Forest`, `XGBoost`, `LightGBM`, `CatBoost` and many more.
  
* Even though as [discussed above](#feature-selection), I was planning to use Lasso embedded method for feature selection during the model training, it turns out that Lasso requires feature scaling, but luckily there are other embedded methods, namely *tree-based* embedded methods, which are `Decision Tree` and `Random Forest` that **require no feature scaling**. <u>Hence *tree-based* models and embedded methods will be used for both feature selection and model training.</u>

* Given that it is the **regression problem**, the MSE, MAE, $R^2$ and RMSE will be used as evaluation metrics for each of the model.

* Since each model requires different hyperparameter, the `GridSearchCV` will be applied separately to each one of them in order to find the `best_params_` for the `n_estimators` parameters of each model.

## $1^{st}$ RandomForest (RF) algorithm

````python
# Grid search for n_estimators

rf_param_options = {"n_estimators": [50, 100, 200, 300]}
rf_md = RandomForestRegressor(random_state=42)
rf_cv = GridSearchCV(rf_md, rf_param_options, cv=5, scoring="r2")
rf_cv.fit(X_train, y_train)

print(f"The best n_estimators hyperparameter for Random Forest: {rf_cv.best_params_}")
````

````python
# fitting RF model

rf_model = RandomForestRegressor(n_estimators=rf_cv.best_params_['n_estimators'], random_state=42)
rf_model.fit(X_train, y_train)
````

````python
# Feature selection based on whether their importance is higher than the median threshold

importance_features_rf = rf_model.feature_importances_
threshold_rf = np.median(importance_features_rf) # setting mediam as threshold
selected_features_rf = X_train.columns[importance_features_rf > threshold_rf]

print(f"Selected features for RF: {list(selected_features_rf)}")
````

````python
# Now train the model on the selected features

rf_model.fit(X_train[selected_features_rf], y_train)
y_pred_rf = rf_model.predict(X_test[selected_features_rf])

print(y_pred_rf) # predictions made with Random Forest
````

## $2^{nd}$ XGBoost (XGB) algorithm

````python
# Grid search for n_estimators

xgb_param_options = {"n_estimators": [50, 100, 200, 300]}
xgb_md = XGBRegressor(random_state=42, objective='reg:squarederror')
xgb_cv = GridSearchCV(xgb_md, xgb_param_options, cv=5, scoring='r2')
xgb_cv.fit(X_train, y_train)

print(f"The best n_estimators for XGB: {xgb_cv.best_params_}")
````

````python
# fitting XGB with the best n_estimators

xgb_model = XGBRegressor(n_estimators=xgb_cv.best_params_['n_estimators'], 
                         random_state=42, objective='reg:squarederror')

xgb_model.fit(X_train, y_train)
````

````python
# Feature selection based on importance

importance_features_xgb = xgb_model.feature_importances_
threshold_xgb = np.median(importance_features_xgb)
selected_features_xgb = X_train.columns[importance_features_xgb > threshold_xgb]

print(f"Selected features for XGB: {list(selected_features_xgb)}")
````

````python
# Fitting the XGB with the selected features

xgb_model.fit(X_train[selected_features_xgb], y_train)
y_pred_xgb = xgb_model.predict(X_test[selected_features_xgb])

print(y_pred_xgb)
````

## $3^{rd}$ LightGBM (LGB) algorithm

````python
# Grid search for n_estimators

lgb_param_options = {"n_estimators": [50, 100, 150, 300]}
lgb_md = LGBMRegressor(random_state=42, min_child_samples=5, verbosity=-1) # -1 to avoid pop up messages
lgb_cv = GridSearchCV(lgb_md, lgb_param_options, cv=5, scoring='r2')
lgb_cv.fit(X_train, y_train)

print(f"The best n_estimators for LGB: {lgb_cv.best_params_}")
````

````python
# fitting the LGB with the best n_estimators

lgb_model = LGBMRegressor(n_estimators=lgb_cv.best_params_['n_estimators'], random_state=42, min_child_samples=5)
lgb_model.fit(X_train, y_train)

````

````python
# Feature selection based on importance

importance_features_lgb = lgb_model.feature_importances_
threshold_lgb = np.median(importance_features_lgb)
selected_features_lgb = X_train.columns[importance_features_lgb > threshold_lgb]

print(f"Selected features for LGB: {list(selected_features_lgb)}")

````

````python
# fitting the model with high performing features

lgb_model.fit(X_train[selected_features_lgb], y_train)
y_pred_lgb = lgb_model.predict(X_test[selected_features_lgb])

print(y_pred_lgb)
````

# Models Evaluation

```text
After feature selection, model training and gettint the predicted values, now those values will be evaluated against the actual true values.
```


````python
# simply receive true and predicted values and compute the loss

def evaluate_model(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, mse, rmse, r2
````

````python
# calculate loss

mae_rf, mse_rf, rmse_rf, r2_rf = evaluate_model(y_test, y_pred_rf)

mae_xgb, mse_xgb, rmse_xgb, r2_xgb = evaluate_model(y_test, y_pred_xgb)

mae_lgb, mse_lgb, rmse_lgb, r2_lgb = evaluate_model(y_test, y_pred_lgb)
````

````python
# comparing performances of each models in dataframe table

results = pd.DataFrame({
     "Model": ["RandomForest", "XGBoost", "LightGBM"],
     "Selected_Features": [len(selected_features_rf), len(selected_features_xgb), len(selected_features_lgb)],
     "MAE": [mae_rf, mae_xgb, mae_lgb],
     "MSE": [mse_rf, mse_xgb, mse_lgb],
     "RMSE": [rmse_rf, rmse_xgb, rmse_lgb],
     "R2": [r2_rf, r2_xgb, r2_lgb]
}, index=[1,2,3])

display(results)
````
